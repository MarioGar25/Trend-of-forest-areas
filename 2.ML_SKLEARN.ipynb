# Tune Model

## Mediterranean Model

```python
import optuna
from optuna.samplers import TPESampler

from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    n_estimators = trial.suggest_int("n_estimator",100,1000)
    max_depth = trial.suggest_int("max_depth", 4, 20)
    max_features = trial.suggest_int("max_features", 5, X_med_train.shape[1])
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 2, 20)
    
    
    model = ExtraTreesClassifier(n_estimators = n_estimators,
                                 max_depth = max_depth,
                                 max_features = max_features,
                                 min_samples_split = min_samples_split,
                                 min_samples_leaf = min_samples_leaf)
    
    score = cross_val_score(estimator = model,
                            X = X_med_train,
                            y = y_med_train,
                            scoring = "accuracy",
                            cv = 5).mean()
    return score
```

```python
study_med_DT = optuna.create_study(study_name = "extratree_med_tune",
                                   direction = "maximize",
                                   sampler = TPESampler())
```

```python
study_med_opt = study_med_DT.optimize(objective, n_trials = 5, n_jobs = -1, show_progress_bar = True )
```

```python
study_med_DT.best_params
```
